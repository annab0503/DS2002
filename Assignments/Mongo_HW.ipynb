## Exercise 3: Aggregation Pipeline

1. Write an aggregation pipeline that calculates the average IMDb rating for each genre (Return the top 5 genres)

```python
# Create the aggregation pipeline
avg_rating_by_genre = collection.aggregate([
  {"$unwind": "$genres"},
  {"$group": {"_id": "$genres", "avg_rating": {"$avg": "$imdb.rating"}}},
  {"$sort": {"avg_rating": -1}},
  {"$limit": 5}
])

# Execute the aggregation pipeline
for genre in avg_rating_by_genre:
    pprint.pprint(genre)
```

2. Write an aggregation pipeline to find the top 5 directors by the average IMDb rating of their movies.


```python
# Create the aggregation pipeline
top_directors = collection.aggregate([
  {"$group": {"_id": "$directors", "avg_rating": {"$avg": "$imdb.rating"}}},
  {"$sort": {"avg_rating": -1}},
  {"$limit": 5}
])

# Execute the aggregation pipeline
for director in top_directors:
    pprint.pprint(director)
```

3. Write an aggregation pipeline to calculate the total number of movies released in each year (Sort the results by the year).


```python
# Create the aggregation pipeline (example code)
movies_per_year = collection.aggregate([
  {"$group": {"_id": "$year", "total_movies": {"$sum": 1}}},
  {"$sort": {"_id": 1}}
])

# Execute the aggregation pipeline
for year in movies_per_year:
    pprint.pprint(year)

# Problem! Some years contain non-integer values.
```

```python
# Execute an aggregation pipeline on the collection
movies_per_year = collection.aggregate([
    # Stage 1: Project a new field called "year"
    {"$project": {
        "year": {   # Convert the year to an integer
        "$toInt": { # Trim any whitespace and clean the year input
        "$trim": {
        "input": {  # Replace occurrences of 'é' in the year string with an empty string
        "$replaceOne": {
        "input": {  # Replace occurrences of 'è' in the year string with an empty string
        "$replaceOne": {  # Convert the year field to a string
        "input": {"$toString": "$year"},
        "find": "é",
        "replacement": "" }},
        "find": "è",  # Specify the character to replace
        "replacement": ""  # Specify the replacement for 'è'
                        }}}}}}},
    # Stage 2: Filter out documents where "year" is None
    {"$match": {"year": {"$ne": None}}},
    # Stage 3: Group the results by year and count the total movies for each year
    {"$group": {"_id": "$year", "total_movies": {"$sum": 1}}},
    # Stage 4: Sort the results by year in ascending order
    {"$sort": {"_id": 1}}
])

# Execute the aggregation pipeline
for year in movies_per_year:
    pprint.pprint(year)

# Problem Solved! This code removes values that contain the non-integer "é"
```

## Exercise 4: Updating and Deleting Documents

1. Write a query to update the IMDb rating of a movie with the title "The Godfather" to 9.5.


```python
# Perform Query
collection.update_one({"title": "The Godfather"}, {"$set": {"imdb.rating": 9.5}})

# Print Results
for movie in collection.find({"title": "The Godfather"}):
    pprint.pprint(movie)
```

2. Write a query to update all movies where the genre is "Horror" and set their IMDb rating to 6.0 if it is currently null.

```python
# Identify Relevant Movies
for movie in collection.find({"genres": "Horror", "imdb.rating": ""}):
    pprint.pprint(movie)
```

```python
# Perform Query
collection.update_many({"genres": "Horror", "imdb.rating": ""}, {"$set": {"imdb.rating": 6.0}})

# Print Results
for movie in collection.find({"genres": "Horror", "imdb.rating": ""}):
    pprint.pprint(movie)
```

3. Write a query to delete all movies that were released before the year 1950.

```python
# Delete all movies that were released before the year 1950.
collection.delete_many({"year": {"$lt": 1950}})

# Print Results
for movie in collection.find({"year": {"$lt": 1950}}):
    pprint.pprint(movie)
```
