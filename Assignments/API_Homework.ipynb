```python
apikey= "Rp5jjOhoEH89QjCtaSr292qWYfZFuYD857g6QQEK"

url = "https://yfapi.net/v6/finance/quote"
url_2 = "https://yfapi.net/v6/finance/quote/target"

querystring = {"symbols": stocks}

headers = {
    'x-api-key': apikey
    }

response = requests.request("GET", url, headers=headers, params=querystring)

# Check the status code
if response.status_code == 200:
    print("API request successful!")
    print(response.text)
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

```python
for stock in stocks:
    querystring = {"symbols": stock.strip()}

    headers = {
        'x-api-key': apikey
    }

    response = requests.request("GET", url, headers=headers, params=querystring)

    stock_json = response.json()
    company_name = stock_json['quoteResponse']['result'][0]["longName"]
    ticker_name = stock_json['quoteResponse']['result'][0]["symbol"]
    current_price = stock_json['quoteResponse']['result'][0]["regularMarketPrice"]
    fifty_two_week_low = stock_json['quoteResponse']['result'][0]["fiftyTwoWeekLow"]
    fifty_two_week_high = stock_json['quoteResponse']['result'][0]["fiftyTwoWeekHigh"]

    print(company_name + " Price: $" + str(current_price) + ", Ticker: " + str(ticker_name) + ", 52 Week Low: $" + str(fifty_two_week_low) + ", 52 Week High: $" + str(fifty_two_week_high))

# can not locate target average price
```

```python
# Create an empty list to store the stock data
data = []

for stock in stocks:
    querystring = {"symbols": stock.strip()}

    headers = {
        'x-api-key': apikey
    }

    response = requests.request("GET", url, headers=headers, params=querystring)

    stock_json = response.json()
    result = stock_json['quoteResponse']['result'][0]

    # Create a dictionary for each stock's information
    stock_info = {
        "Ticker": result["symbol"],
        "Company Name": result["longName"],
        "Current Price": result["regularMarketPrice"],
        "52 Week Low": result["fiftyTwoWeekLow"],
        "52 Week High": result["fiftyTwoWeekHigh"]
    }

    # Append the dictionary to the list
    data.append(stock_info)

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
# Save the df as a csv
df.to_csv('stocks_data.csv', index=False)
```
